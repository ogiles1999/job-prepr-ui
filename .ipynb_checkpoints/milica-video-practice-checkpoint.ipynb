## Converting a video to frames

```python
import cv2
import matplotlib.pyplot as plt

vidcap = cv2.VideoCapture('video-test-milica.mp4')
success,image = vidcap.read()
images=[]

while success:   
    success, image = vidcap.read()
    images.append(image)

plt.imshow(images[50])
```

```python
len(images)
```

## Extracting the face

```python
from skimage import color
from PIL import Image
import numpy as np

images_just_faces=[]
for image in images:
    try:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
 
        faces = face_cascade.detectMultiScale(
        image,
        scaleFactor=1.3,
        minNeighbors=3,
        minSize=(30, 30)
                )
        for (x, y, w, h) in faces:
            cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
            just_face = image[y:y + h, x:x + w]
            just_face=np.array(Image.fromarray(just_face).resize((48,48)))
            images_just_faces.append(just_face)
    except:
        pass
```

```python
print(f"We have {len(images_just_faces)} images in total. Showing a sample picture. The dimensions are: {images_just_faces[10].shape}.")
plt.imshow(images_just_faces[10], cmap="gray");
```

# Random questions

```python
import pandas as pd
import random
```

```python
def load_questions():
    questions = pd.read_csv('interview_questions.csv',
                            header=0,
                            names=["Area", "Question"],
                            on_bad_lines='skip',
                            delimiter=";")
    return questions
```

```python
def get_rand_question(questions, job_name):
    questions=questions[questions["Area"]==job_name]
    no_of_q=questions.shape[0]
    index=random.randint(0, no_of_q-1)

    return questions["Question"].iloc[index]
```

```python
questions=load_questions()
job_name = "Programming"
question=get_rand_question(questions, job_name)
question
```
